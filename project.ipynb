#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

Print both values for each variable

```python
# TODO
# this CSV file contains semicolons instead of comas as separator
df = pd.read_csv('assets/real_estate.csv', sep=';')

south_df = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]
filtered_df = df[df['level5'].isin(south_df)]

priceM = filtered_df.price.mean()
priceV = filtered_df.price.var()
print('Price mean and variance', priceM, priceV)

SurM = filtered_df.surface.mean()
SurV = filtered_df.surface.var()
print('Surface mean and variance', SurM, SurV)

batM = filtered_df.bathrooms.mean()
batV = filtered_df.bathrooms.var()
print('Bathroom mean and variance', batM, batV)

roomM = filtered_df.rooms.mean()
roomV = filtered_df.rooms.var()
print('Rooms mean and variance', roomM, roomV)
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python

# TODO
import matplotlib.pyplot as plt 
import pandas as pd


df = pd.read_csv('assets/real_estate.csv', sep=';')
south_df = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]
filtered_df = df[df['level5'].isin(south_df)]

fig, axs = plt.subplots(len(south_df), 1, figsize=(8, 6 * len(south_df)))

for i, location in enumerate(south_df):
    filtered_df = df[df['level5'] == location]
    normalized_df = (filtered_df['price']-filtered_df['price'].min())/(filtered_df['price'].max()-filtered_df['price'].min())
    axs[i].hist(normalized_df, bins=30, color='skyblue', edgecolor='black')
    axs[i].set_xlabel('Price')
    axs[i].set_ylabel('Frequency')
    axs[i].set_title(f'Histogram for {location}')
    
    
plt.tight_layout()
plt.show()
```
